#Photosystem II

Photolysis of water 2H2O-->4H+ + 4e + O2

```python
oxygen = vlx.Molecule.read_smiles("O=O")
oxygen.set_multiplicity(3)
oxygen.show(atom_indices=True)

basis = vlx.MolecularBasis.read(oxygen, 'def2-SVP')

scf_drv = vlx.ScfUnrestrictedDriver()
scf_drv.xcfun = "blyp"

scf_drv.compute(oxygen, basis)

energy_oxygen = scf_drv.get_scf_energy()
print(energy_oxygen)
```

energy of oxygen is -150.19242078692614 Hartee

```python
protons = vlx.Molecule.read_smiles("[H+].[H+].[H+].[H+]")
protons.show(atom_indices=True)
#SCF needs elcetrons and therefore energy of protons becomes 0 Hartee
energy_proton=0.0 #hartree
energy_electron=0.0 #SCF doesn't work on free electrons 
```

```python

# Molecule
water = vlx.Molecule.read_smiles("O")
water.set_multiplicity(1)  # singlet!
water.show(atom_indices=True)

# Basis
basis = vlx.MolecularBasis.read(water, 'def2-SVP')

# SCF driver
scf_water = vlx.ScfRestrictedDriver()
scf_water.xcfun = "blyp"

# Compute SCF
scf_water.compute(water, basis)

# Get energy
energy_water = scf_water.get_scf_energy()
print("Water SCF energy (Hartree):", energy_water)
```

Water SCF energy (Hartree): -76.33678505510426

```python
#Reaction energy photolysis of water 
reactionE_1=(4*energy_proton+4*energy_electron+energy_oxygen)-2*energy_water 
print("Reaction energy photolysis (Hartree):", reactionE_1)

```
