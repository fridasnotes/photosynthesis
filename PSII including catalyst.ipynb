```python
import veloxchem as vlx

# --------------------------
# Reactant system: Mn2O2 + 2 H2O molecules
# --------------------------
reactant_xyz = """10
Mn2O2 + 2 H2O minimal model
Mn  0.0  0.0  0.0
Mn  2.0  0.0  0.0
O   1.0  1.0  0.0
O   1.0 -1.0  0.0
O   0.5  0.0  1.5
H   0.5  0.96 1.9
H   0.5 -0.96 1.9
O   1.5  0.0  1.5
H   1.5  0.96 1.9
H   1.5 -0.96 1.9
"""

reactant_system = vlx.Molecule.read_xyz_string(reactant_xyz)
reactant_system.set_multiplicity(5)  # high-spin due to Mn

# --------------------------
# Product system: Mn2O2 + O2 molecule
# --------------------------
product_xyz = """6
Mn2O2 + O2 minimal model
Mn  0.0  0.0  0.0
Mn  2.0  0.0  0.0
O   1.0  1.0  0.0
O   1.0 -1.0  0.0
O   0.0  0.0  2.0
O   0.0  0.0 -2.0
"""

product_system = vlx.Molecule.read_xyz_string(product_xyz)
product_system.set_multiplicity(5)

# --------------------------
# Basis sets
# --------------------------
basis_react = vlx.MolecularBasis.read(reactant_system, "sto-3g")
basis_prod  = vlx.MolecularBasis.read(product_system, "sto-3g")

# --------------------------
# SCF calculations
# --------------------------
scf_react = vlx.ScfUnrestrictedDriver()
scf_react.xcfun = "blyp"
scf_react.compute(reactant_system, basis_react)
E_react = scf_react.get_scf_energy()

scf_prod = vlx.ScfUnrestrictedDriver()
scf_prod.xcfun = "blyp"
scf_prod.compute(product_system, basis_prod)
E_prod = scf_prod.get_scf_energy()

# H+ and e- → 0 Hartree
E_Hplus = 0.0
E_e = 0.0

# --------------------------
# Reaction energy
# --------------------------
delta_E = (E_prod + 4*E_Hplus + 4*E_e) - E_react
print("Reaction energy with catalyst (Hartree):", delta_E)
print("Reaction energy with catalyst (kJ/mol):", delta_E * 2625.5)
```
